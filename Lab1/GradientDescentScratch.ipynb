Implement (from scratch!) gradient descent to find
minimum value of a given function f(x) and its first order
derivative f_()
• Print the intermediate iterative steps, similar to the previous
table (time, x, f(x))
• Try experimenting with the previous example f(x) = x^2

```python
def f(x):
    """Target function: f(x) = x^2"""
    return x ** 2


def f_(x):
    """First-order derivative: f'(x) = 2x"""
    return 2 * x
```

```python
def gradient_descent(f, f_, x_init, learning_rate, max_iter=2000, threshold=0.0001):
    """
    Minimise f(x) by stepping against the gradient.
    Update rule:  x  <-  x  -  r * f'(x)
    """
    x = float(x_init)
    fx   = f(x)
    counter = 1
    while fx > threshold:
        grad = f_(x)
        x = x - learning_rate * grad
        fx = f(x)
        print(f"{counter}. x: {x} - fx: {fx}")
        counter += 1
```

```python
learning_rate = 0.01
x = 2 # random number
gradient_descent(f, f_, x, learning_rate)

```
