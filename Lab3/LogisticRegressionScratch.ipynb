
# Logistic Regression from Scratch using Gradient Descent



```python
import csv
import math
```

```python
def load_csv(path: str):
    X, y = [], []
    with open(path, newline="") as f:
        reader = csv.reader(f)
        header = next(reader, None)
        for row in reader:
            x1, x2, label = float(row[0]), float(row[1]), float(row[2])
            X.append([1.0, x1, x2])
            y.append(label)
    return X, y
```

```python
def sigmoid(z: float) -> float:
    return 1.0 / (1.0 + math.exp(-z))
```

```python
def dot(a: list, b: list) -> float:
    # dot_product = a1*b1+a2*b2+a3*b3+...+an*bn
    return sum(ai * bi for ai, bi in zip(a, b))


def predict_prob(w: list, x: list) -> float:
    return sigmoid(dot(w, x))
```

```python
def binary_cross_entropy(w: list, X: list, y: list) -> float:

    n = len(y)
    total = 0.0
    for xi, yi in zip(X, y):
        p = predict_prob(w, xi)
        total += yi * math.log(p) + (1 - yi) * math.log(1 - p)
    return -total / n
```

```python
def compute_gradient(w: list, X: list, y: list) -> list:
    n = len(y)
    grad = [0.0] * len(w)
    for xi, yi in zip(X, y):
        error = predict_prob(w, xi) - yi
        for j in range(len(w)):
            grad[j] += error * xi[j]
    return [g / n for g in grad]
```

```python
def gradient_descent(X: list, y: list,
                     lr: float = 0.1,
                     tol: float = 0.2):
    w = [0.0, 1.0, 2.0] # initialise weights to zero

    print("=" * 65)
    print(f"{'Iter':>6}  {'Loss':>14}  {'w0':>10}  {'w1':>10}  {'w2':>10}")
    print("=" * 65)
    loss = 100
    epoch = 1
    while loss > tol:
        grad = compute_gradient(w, X, y)
        loss = binary_cross_entropy(w, X, y)
        print(f"{epoch:>6}  {loss:>14.6f}  {w[0]:>10.5f}  {w[1]:>10.5f}  {w[2]:>10.5f}")
        w = [wj - lr * gj for wj, gj in zip(w, grad)]

        if loss < tol:
            print(f"\nConverged at iteration {epoch}  (||grad|| = {loss:.2e})")
            break

        epoch += 1
    print("=" * 65)
    return w
```

```python
def evaluate(w: list, X: list, y: list):
    correct = sum(
        1 for xi, yi in zip(X, y)
        if round(predict_prob(w, xi)) == yi
    )
    acc = correct / len(y) * 100
    print(f"\nTraining accuracy : {correct}/{len(y)}  ({acc:.1f} %)")
```

```python
def main():
    csv_path = "loan2.csv"

    X, y = load_csv(csv_path)

    w = gradient_descent(X, y, lr=0.1)

    print(f"\n{'─'*40}")
    print("  Final weights:")
    print(f"    w0 (bias) = {w[0]:.6f}")
    print(f"    w1        = {w[1]:.6f}")
    print(f"    w2        = {w[2]:.6f}")
    print(f"{'─'*40}")

    evaluate(w, X, y)

if __name__ == "__main__":
    main()
```
